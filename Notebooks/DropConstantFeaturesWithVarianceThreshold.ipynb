```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.feature_selection import VarianceThreshold
```

Creating the dataset

```python
data = pd.DataFrame(
    {
        'A':[1,2,4,1,2,4],
        'B':[4,5,6,7,8,9],
        'C':[0,0,0,0,0,0], 
        'D':[1,1,1,1,1,1], 
    }
)
```

We can see that column C and D are constant column where values are not changing

```python
data.head()
```

```python
thresh = VarianceThreshold(threshold=0.2)
thresh.fit(data)
```

```python
thresh.feature_names_in_
```

```python
thresh.get_support()
```

```python
selected_columns = data.columns[thresh.get_support()]
```

```python
selected_columns
```

```python
data = data[selected_columns.to_list()]
```

```python
data.head()
```
